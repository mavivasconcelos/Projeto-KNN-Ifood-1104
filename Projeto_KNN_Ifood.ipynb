## Ponto será comparado com demais e terá sua moda de classe calculada

```python
# Função para calcular distância entre dois pontos
# A intenção é que a coordenada seja definida em outra função que escolha o ponto
# Com coord1 sendo um ponto escolhido (ponto_escolhido) e coord2 sendos os demais pontos Y
def calcular_distancia (coord1, coord2):

  for i in data + no_class:
    calculo_distancia = ((coord1[2][0]-coord2[2][0])**2 +
                        (coord1[2][1]-coord2[2][1])**2 +
                        (coord1[2][2]-coord2[2][2])**2 +
                        (coord1[2][3]-coord2[2][3])**2)**0.5
  return calculo_distancia

# Função para calcular distância entre o ponto escolhido e os demais pontos
def calcular_distancias_totais (ponto_escolhido, data, no_class):
  distancias_totais = []
  for i in data + no_class:
    if i != ponto_escolhido:
      calculo_distancias_totais = calcular_distancia (ponto_escolhido, i)
# É adicionada a nova lista tando os item do ponto_escolhido quanto a distância total
# por vias de análise de veracidade e comparação do resultado final indivudual do ponto teste
      distancias_totais.append([i, calculo_distancias_totais])
  distancias_totais_ordenadas = sorted(distancias_totais, key=lambda ponto: ponto[1])
  return distancias_totais_ordenadas

def selecionar_vizinhos (distancias_totais_ordenadas, k):
  knn_vizinhos = distancias_totais_ordenadas[:k]
  return knn_vizinhos

# De acordo com knn_vizinhos adicionar em dicionario_NPS os NPS que são diferentes de ''
def NPS_vizinhos (knn_vizinhos):
  dicionario_NPS = {}
  for vizinho in knn_vizinhos:
    if vizinho[0][1] != '':
      classe = vizinho[0][1]
      if classe in dicionario_NPS:
        dicionario_NPS[classe] += 1
      else:
        dicionario_NPS[classe] = 1
  contagem_maxima = 0
  return dicionario_NPS

# Calcular moda de dicionario_NPS
def calcular_moda(dicionario_NPS):
    moda = None
    max_contagem = 0

    for classe, contagem in dicionario_NPS.items():
        if contagem > max_contagem:
            moda = classe
            max_contagem = contagem
    return moda

# Calcular estimativa para cada item da lista no_class
def estimativa_no_class_vizinhos (moda, no_class):
# Preciso criar uma nova lista para armazenar novos itens

  no_class_vizinhos = []
# Preciso calcular os vizinhos de cada item
# Calcular moda dos vizinhos de cada item
# Adicionar na nova lista o ponto com a classe '' alterada pela moda
  for usuario in no_class:
    moda = calcular_moda(NPS_vizinhos (selecionar_vizinhos (calcular_distancias_totais (usuario, data, no_class), k)))
    item_modificado = (usuario[0], moda, usuario[2])
    no_class_vizinhos.append(item_modificado)
  return no_class_vizinhos


# Definir ponto para exemplo individual de cálculo
# Coloquei a possibilidade de escolher e calcular KNN e a moda indivudualmente para poder analisar a assertividade do resultado
ponto_escolhido = [66175672425, '', (51.00, 35.00, 14.00, 30.0)]


# Definir parâmetros para cálculos e comparações
k = 5
distancias_totais_ordenadas = calcular_distancias_totais (ponto_escolhido, data, no_class)
knn_vizinhos = selecionar_vizinhos (distancias_totais_ordenadas, k)
dicionario_NPS = NPS_vizinhos (knn_vizinhos)
moda = calcular_moda (dicionario_NPS)
no_class_vizinhos = estimativa_no_class_vizinhos (moda, no_class)


# Print ordena os K vizinhos de determinado ponto.
print (selecionar_vizinhos (distancias_totais_ordenadas, k))


#Print estimativa de NPS de determinado ponto
print (f"Estimativa de NPS do Cliente de CPF {ponto_escolhido[0]}:\n\n{moda}")
```
